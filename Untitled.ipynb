```python
skew_vals = data.skew()
# Positive skew (> 0)
pos_skew = skew_vals[skew_vals > 0]

# Negative skew (< 0)
neg_skew = skew_vals[skew_vals < 0]

# Approximately zero skew (between -0.1 and 0.1)
zero_skew = skew_vals[(skew_vals >= -0.1) & (skew_vals <= 0.1)]
print("Positive Skewed Columns:\n", pos_skew)

print("\nNegative Skewed Columns:\n", neg_skew)

print("\nApproximately Zero Skew Columns:\n", zero_skew)
```

```python
pos_count = (skew_vals > 0).sum()
neg_count = (skew_vals < 0).sum()
zero_count = ((skew_vals >= -0.1) & (skew_vals <= 0.1)).sum()

print("Positive Skew Count:", pos_count)
print("Negative Skew Count:", neg_count)
print("Approximately Zero Skew Count:", zero_count)
```

```python
# Log transform
X = np.log1p(X)

print("X shape:", X.shape)   # (36043, 98)
print("y shape:", y.shape)
```

```python
# Use SelectKBest to score all 98 features and pick top 10
selector = SelectKBest(mutual_info_regression, k=15)
selector.fit(X, y)

# Get the 10 selected feature names
selected_15 = list(X.columns[selector.get_support()])

print("Top 10 Selected Features:")
for i, f in enumerate(selected_10, 1):
    print(f"  {i}. {f}")
```

```python
# Keep only the 10 selected features
X_15 = X[selected_15]

print("New X shape:", X_15.shape)  # (36043, 10)
X_15.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_15, y, test_size=0.2, random_state=42
)

print("Train:", X_train.shape)
print("Test: ", X_test.shape)
```

```python
model = Pipeline([
    ("Scaler", RobustScaler()),
    ("RF",     RandomForestRegressor(random_state=42))
])
model
```

```python
model.fit(X_train, y_train)
print("Model trained!")
```

```python
y_pred = model.predict(X_test)

print(f"R2 Score : {r2_score(y_test, y_pred):.4f}")
print(f"MAE      : {mean_absolute_error(y_test, y_pred):,.2f}")
print(f"RMSE     : {mean_squared_error(y_test, y_pred)**0.5:,.2f}")
```

```python
joblib.dump({"model": model, "features": selected_15}, "model_15.pkl")
print("Saved model_15.pkl")
print("Features saved:", selected_15)
```
